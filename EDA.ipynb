## No. of words per sentence

```python
def countWords(sentences):
    countWords = []
    for sentence in sentences:
        countWords.append(len(sentence.split(" ")))
        
    return countWords[:]

numberOfWords = countWords(df['text'].values[:])
numberOfWords[:5]
```

```python
plt.figure(figsize=(10,5))
plt.xticks(np.arange(0, max(numberOfWords)+1, 100.0))
plt.hist(numberOfWords,normed=1,cumulative=True, histtype='step', bins = 20)
plt.show()
```

This implies approx 90% of sentences contains 200 words or less. This data is very important and will be used when we actually design out LSTM system
