```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
pd.set_option('display.max_rows', 500)
pd.set_option("display.max_columns", 10)
pd.set_option('display.max_colwidth', -1)

%matplotlib inline
```

```python
df = pd.read_csv("./parsed_xmlAsCSV.csv")
print ("No of rows read : ", len(df))
df.head()
```

```python
df.hist(figsize=(15,10))
```

```python
c=Counter(df['nationality'])
print (c)
print ("No. of nationalities : ", len(c))
plt.figure(figsize=(10,7))
df['nationality'].value_counts().plot(kind='bar')
plt.plot()
```

```python
df = df.reindex(np.random.permutation(df.index)).reset_index(drop=True)
df.head()
```

##  Partioning

Partioning dataframe into 25 separate DF

```python
lenghtPerDF = int(len(df)/25)
for i in range(0, 25):
    tempDf = df[i*lenghtPerDF:i*lenghtPerDF+lenghtPerDF]
    tempDf.to_csv("./Part"+str(i+1)+".csv", index=False)
```
